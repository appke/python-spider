```python
import requests, re
from bs4 import BeautifulSoup
import time


for i in range(10):
    url = 'http://maoyan.com/board/4?offset=%d'%(i*10)
#     print(url)
    time.sleep(1.5)
    res = requests.get(url)
    res.encoding = 'utf-8'
# soup = BeautifulSoup(res.text, 'lxml')
# divs = soup.find_all('div', class_='movie-item-info')


    divs = re.findall(r'<div class="movie-item-info">(.*?)</div>', res.text, re.S)

    for film in divs:
        name = re.search(r'<p class="name"><a.*?>(.*?)</a></p>', film).group(1)
        star = re.search(r'<p class="star">(.*?)</p>', film, re.S).group(1).strip()
        releasetime = re.search(r'<p class="releasetime">(.*?)</p>', film).group(1)
        print('《'+name+'》', star, releasetime)
        with open(r'dy.txt','a+') as f:
            f.write('《'+name+'》'+'\t'+star+'\t'+releasetime+'\n') 
```
